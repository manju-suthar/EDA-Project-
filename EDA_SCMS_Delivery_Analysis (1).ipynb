3. Data Wrangling

Data Warngling Code

```python
# Data Wrangling

# 1. Clean Numeric Columns (Handling 'See ASN' and 'Freight Included' text)
# We convert them to 'coerce' which turns non-numeric text into NaN
cols_to_fix = ['Weight (Kilograms)', 'Freight Cost (USD)', 'Line Item Insurance (USD)']
for col in cols_to_fix:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 2. Impute (Fill) Missing Values
# We use Median for costs/weights because it is not affected by outliers
df['Weight (Kilograms)'].fillna(df['Weight (Kilograms)'].median(), inplace=True)
df['Freight Cost (USD)'].fillna(df['Freight Cost (USD)'].median(), inplace=True)

# We use Mode (most frequent) for categorical columns
df['Shipment Mode'].fillna(df['Shipment Mode'].mode()[0], inplace=True)
df['Line Item Insurance (USD)'].fillna(0, inplace=True)
df['Dosage'].fillna('Not Specified', inplace=True)

# 3. Convert Date Columns to Datetime Objects
df['Scheduled Delivery Date'] = pd.to_datetime(df['Scheduled Delivery Date'], errors='coerce')
df['Delivered to Client Date'] = pd.to_datetime(df['Delivered to Client Date'], errors='coerce')

# 4. Feature Engineering: Create 'Delivery Delay'
# This is the most valuable column for your internship analysis
df['Delivery Delay'] = (df['Delivered to Client Date'] - df['Scheduled Delivery Date']).dt.days

# 5. Drop Unnecessary Columns (Optional)
# Columns with too many unique IDs or constant values don't help analysis
# df.drop(['ID', 'Managed By'], axis=1, inplace=True)

print("Data Wrangling Complete!")
print(f"Dataset now has {df.isnull().sum().sum()} total missing values.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style
sns.set_theme(style="white")

# 1. Selecting only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

# 2. Calculating the correlation matrix
corr_matrix = numeric_df.corr()

# 3. Setting up the matplotlib figure
plt.figure(figsize=(12, 10))

# 4. Generating a mask for the upper triangle (optional, for cleaner look)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# 5. Drawing the heatmap
sns.heatmap(corr_matrix,
            mask=mask,
            annot=True,
            fmt=".2f",
            cmap='coolwarm',
            square=True,
            linewidths=.5,
            cbar_kws={"shrink": .8})

plt.title('Correlation Heatmap of Supply Chain Metrics', fontsize=16)
plt.show()
```
